```python
import pandas as pd
import tensorflow as tf
from transformers import DistilBertTokenizer, TFDistilBertModel
from keras.layers import Dense, Concatenate, Input, Lambda
from keras.models import Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.optimizers import Adam
import matplotlib.pyplot as plt


# 1. 데이터 전처리
def preprocess(file_path, features, label_column="Type", test_size=0.2, random_state=35):
    df = pd.read_csv(file_path, encoding='utf-8')
    columns = [label_column] + features
    df[features] = df[features].astype(str)
    df = df[columns]

    # 레이블 인코딩
    label_encoder = LabelEncoder()
    df['label_encoded'] = label_encoder.fit_transform(df[label_column])
    num_classes = len(label_encoder.classes_)

    # Train/Test Split
    train_df, val_df = train_test_split(df, test_size=test_size, random_state=random_state)
    return train_df, val_df, num_classes, label_encoder


# 2. DistilBERT 토크나이저 및 인코딩
def tokenize_features(df, features, tokenizer, max_length=512):
    encodings = {}
    for feature in features:
        encoding = tokenizer(
            list(df[feature]),
            max_length=max_length,
            padding="max_length",
            truncation=True,
            return_tensors="tf"
        )
        encodings[feature] = (encoding["input_ids"], encoding["attention_mask"])
    return encodings


# 3. BERT 임베딩 레이어 정의
def bert_embedding_layer(input_ids, attention_mask, bert_model):
    return Lambda(
        lambda x: bert_model(x[0], attention_mask=x[1]).last_hidden_state[:, 0, :],  # [CLS] 토큰 임베딩 추출
        output_shape=(768,)
    )([input_ids, attention_mask])


# 4. 모델 정의 및 컴파일
def create_model(features, num_classes, bert_model, learning_rate=3e-5):
    feature_embeddings = []
    model_inputs = []
    for feature in features:
        input_ids = Input(shape=(512,), dtype=tf.int32, name=f"{feature.replace(' ', '_')}_input_ids")
        attention_mask = Input(shape=(512,), dtype=tf.int32, name=f"{feature.replace(' ', '_')}_attention_mask")

        # 임베딩 생성
        feature_embedding = bert_embedding_layer(input_ids, attention_mask, bert_model)
        feature_embeddings.append(feature_embedding)

        # 모델 입력에 추가
        model_inputs.extend([input_ids, attention_mask])

    # 모든 임베딩 결합 및 분류 레이어 추가
    concatenated_features = Concatenate()(feature_embeddings)
    output = Dense(num_classes, activation='softmax')(concatenated_features)

    # 모델 정의
    model = Model(inputs=model_inputs, outputs=output)

    # 컴파일
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model


# 5. 모델 학습
def fit_model(model, train_inputs, train_labels, val_inputs, val_labels, epochs=100, batch_size=4):
    history = model.fit(
        train_inputs,
        train_labels,
        validation_data=(val_inputs, val_labels),
        epochs=epochs,
        batch_size=batch_size
    )
    return history


def plot_training_history(history):
    """
    Plots training and validation loss over epochs from a model's training history.
    
    Parameters:
    - history: The history object returned by model.fit(), containing training and validation loss.
    """
    # Extract loss values for each epoch
    epochs = range(1, len(history.history['loss']) + 1)
    train_loss = history.history['loss']
    val_loss = history.history['val_loss']

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_loss, 'r-', label='Train Loss')  # Red for training loss
    plt.plot(epochs, val_loss, 'b-', label='Validation Loss')  # Blue for validation loss
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.grid(True)
    plt.show()


# 6. 메인 실행 함수
def main(file_path):
    # 설정
    features = [
        "Memory", "Language", "Orientation",
        "Judgment and Problem Solving", "Social Activities", "Home and Hobbies",
        "Daily Living", "Personality and Behavior"
    ]

    # 데이터 전처리
    train_df, val_df, num_classes, label_encoder = preprocess(file_path, features)

    # BERT 모델 및 토크나이저 초기화
    tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
    bert_model = TFDistilBertModel.from_pretrained('distilbert-base-uncased')
    #bert_model.trainable = False  # BERT 모델의 파라미터를 고정

    # 인코딩 생성
    train_encodings = tokenize_features(train_df, features, tokenizer)
    val_encodings = tokenize_features(val_df, features, tokenizer)

    # 레이블 준비
    train_labels = tf.convert_to_tensor(train_df['label_encoded'], dtype=tf.int32)
    val_labels = tf.convert_to_tensor(val_df['label_encoded'], dtype=tf.int32)

    # 모델 정의 및 컴파일
    model = create_model(features, num_classes, bert_model)
    model.summary()

    # 입력 데이터 준비
    train_inputs = {}
    val_inputs = {}
    for feature in features:
        feature_key = feature.replace(" ", "_")
        train_inputs[f"{feature_key}_input_ids"] = train_encodings[feature][0]
        train_inputs[f"{feature_key}_attention_mask"] = train_encodings[feature][1]
        val_inputs[f"{feature_key}_input_ids"] = val_encodings[feature][0]
        val_inputs[f"{feature_key}_attention_mask"] = val_encodings[feature][1]

    # 모델 학습
    history = fit_model(model, train_inputs, train_labels, val_inputs, val_labels)
    plot_training_history(history)

    return model, history


# 실행
# file_path에 CSV 파일 경로를 입력하세요.
model, history = main("template.csv")
```

```python
import pandas as pd
import tensorflow as tf
from transformers import DistilBertTokenizer, TFDistilBertModel
from keras.layers import Dense, Input, LSTM
from keras.models import Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.optimizers import Adam
import matplotlib.pyplot as plt


# GPU 설정
def setup_gpu():
    gpus = tf.config.experimental.list_physical_devices('GPU')
    if gpus:
        try:
            tf.config.experimental.set_visible_devices(gpus[0], 'GPU')
            tf.config.experimental.set_memory_growth(gpus[0], True)
            print("Using GPU:", gpus[0].name)
        except RuntimeError as e:
            print(e)


# 1. 데이터 전처리 및 Train/Test Split
def preprocess(file_path, features, label_column="Type", test_size=0.2, random_state=35):
    df = pd.read_csv(file_path, encoding='utf-8')
    columns = [label_column] + features
    df[features] = df[features].astype(str)
    df = df[columns]
    
    # 레이블 인코딩
    label_encoder = LabelEncoder()
    df['label_encoded'] = label_encoder.fit_transform(df[label_column])
    num_classes = len(label_encoder.classes_)
    
    # Train/Test Split
    train_df, val_df = train_test_split(df, test_size=test_size)
    return train_df, val_df, num_classes


# 2. 토크나이저 및 임베딩 생성
def tokenize_embedding(df, features, tokenizer, bert_model, max_length=512):
    embeddings = []
    for feature in features:
        encoding = tokenizer(
            list(df[feature]),
            max_length=max_length,
            padding="max_length",
            truncation=True,
            return_tensors="tf"
        )
        inputs = [encoding["input_ids"], encoding["attention_mask"]]
        feature_embeddings = bert_model(inputs)[0][:, 0, :]  # CLS token
        embeddings.append(feature_embeddings)
    return tf.stack(embeddings, axis=1) 


# 3. 모델 정의 및 컴파일
def create_model(input_shape, num_classes, learning_rate=3e-5):
    input_layer = Input(shape=input_shape)
    #x = LSTM(128)(x)
    #x = LSTM(256, return_sequences=True)(input_layer) # 삭제 가능.
    x = LSTM(128, return_sequences=False)(x)
    dense_output = Dense(64, activation='relu')(x)
    final_output = Dense(num_classes, activation='softmax')(dense_output)
    
    model = Model(inputs=input_layer, outputs=final_output)
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model


# 4. 모델 학습
def fit_model(model, train_embeddings, train_labels, val_embeddings, val_labels, epochs=100, batch_size=4):
    history = model.fit(
        train_embeddings,
        train_labels,
        validation_data=(val_embeddings, val_labels),
        epochs=epochs,
        batch_size=batch_size
    )
    return history


def plot_training_history(history):
    """
    Plots training and validation loss over epochs from a model's training history.
    
    Parameters:
    - history: The history object returned by model.fit(), containing training and validation loss.
    """
    # Extract loss values for each epoch
    epochs = range(1, len(history.history['loss']) + 1)
    train_loss = history.history['loss']
    val_loss = history.history['val_loss']

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_loss, 'r-', label='Train Loss')  # Red for training loss
    plt.plot(epochs, val_loss, 'b-', label='Validation Loss')  # Blue for validation loss
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.grid(True)
    plt.show()


# 메인 실행 함수
def main(file_path):
    setup_gpu()

    # 설정
    features = [
        "Memory", "Language", "Orientation",
        "Judgment and Problem Solving", "Social Activities", "Home and Hobbies",
        "Daily Living", "Personality and Behavior"
    ]

    # 데이터 전처리
    train_df, val_df, num_classes = preprocess(file_path, features)

    # BERT 모델 및 토크나이저 초기화
    tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
    bert_model = TFDistilBertModel.from_pretrained('distilbert-base-uncased')
    #bert_model.trainable = False  # BERT 모델의 파라미터를 고정

    # 임베딩 생성
    train_embeddings = tokenize_embedding(train_df, features, tokenizer, bert_model)
    val_embeddings = tokenize_embedding(val_df, features, tokenizer, bert_model)

    # 레이블 준비
    train_labels = tf.convert_to_tensor(train_df['label_encoded'], dtype=tf.int32)
    val_labels = tf.convert_to_tensor(val_df['label_encoded'], dtype=tf.int32)

    # 모델 정의 및 컴파일
    model = create_model(input_shape=(len(features), 768), num_classes=num_classes)
    model.summary()

    # 모델 학습
    history = fit_model(model, train_embeddings, train_labels, val_embeddings, val_labels)

    plot_training_history(history)

    return model, history

# file_path에 CSV 파일 경로를 입력
model, history = main("template.csv")
```

```python
import pandas as pd
import tensorflow as tf
from transformers import DistilBertTokenizer, TFDistilBertModel
from keras.layers import Input, Dense, Concatenate, Layer, MultiHeadAttention, LayerNormalization
from keras.models import Model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.optimizers import Adam
import matplotlib.pyplot as plt


class DistilBERTEmbeddingLayer(Layer):
    def __init__(self, bert_model, **kwargs):
        super(DistilBERTEmbeddingLayer, self).__init__(**kwargs)
        self.bert_model = bert_model  # Reuse the passed DistilBERT model
        # self.bert_model.trainable = False

    def call(self, inputs):
        input_ids, attention_mask = inputs
        output = self.bert_model(input_ids=input_ids, attention_mask=attention_mask)
        return output.last_hidden_state[:, 0, :]  # Extract [CLS] token embedding
    
    def compute_output_shape(self, input_shape):
        return (input_shape[0], 768)  # [CLS] 토큰 임베딩의 출력 크기

    def get_config(self):
        # This is required for model serialization
        config = super(DistilBERTEmbeddingLayer, self).get_config()
        # We cannot directly serialize bert_model, so we exclude it
        return config

    @classmethod
    def from_config(cls, config):
        # Recreate the BERT model when loading the layer
        bert_model = TFDistilBertModel.from_pretrained('distilbert-base-uncased')
        # bert_model = TFAutoModel.from_pretrained("microsoft/MiniLM-L12-H384-uncased")
        # bert_model.trainable = False
        return cls(bert_model=bert_model, **config)

# 1. 데이터 전처리
def preprocess(file_path, features, label_column="Type", test_size=0.2, random_state=35):
    df = pd.read_csv(file_path, encoding='utf-8')
    columns = [label_column] + features
    df[features] = df[features].astype(str)
    df = df[columns]

    # 레이블 인코딩
    label_encoder = LabelEncoder()
    df['label_encoded'] = label_encoder.fit_transform(df[label_column])
    num_classes = len(label_encoder.classes_)

    # Train/Test Split
    train_df, val_df = train_test_split(df, test_size=test_size, random_state=random_state)
    return train_df, val_df, num_classes, label_encoder


def tokenize_features(df, features, tokenizer, max_length=512):
    encodings = {}
    for feature in features:
        encoding = tokenizer(
            list(df[feature]),
            max_length=max_length,
            padding="max_length",
            truncation=True,
            return_tensors="tf"
        )
        encodings[feature] = (encoding["input_ids"], encoding["attention_mask"])
    return encodings



# 4. 모델 정의 및 컴파일
def create_model(features, num_classes, bert_model, learning_rate=3e-5):
    feature_embeddings = []
    model_inputs = []
    shared_embedding_layer = DistilBERTEmbeddingLayer(bert_model=bert_model)

    for feature in features:
        input_ids = Input(shape=(512,), dtype=tf.int32, name=f"{feature.replace(' ', '_')}_input_ids")
        attention_mask = Input(shape=(512,), dtype=tf.int32, name=f"{feature.replace(' ', '_')}_attention_mask")

        # BERT 임베딩 생성
        feature_embedding = shared_embedding_layer([input_ids, attention_mask])
        feature_embeddings.append(feature_embedding)

        # 모델 입력에 추가
        model_inputs.extend([input_ids, attention_mask])

    # 모든 임베딩 결합 및 분류 레이어 추가
    concatenated_features = Concatenate()(feature_embeddings)

    output = Dense(128, activation='relu')(concatenated_features)
    #output = tf.expand_dims(output, axis=1)

    # 모델 정의
    model = Model(inputs=model_inputs, outputs=output)

    attention_output = MultiHeadAttention(num_heads=4, key_dim=32)(output, output)
    attention_output = LayerNormalization()(attention_output + output)
    output = Dense(64, activation='relu')(attention_output)
    output = Dense(32, activation='relu')(output)

    # 컴파일
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model


# 5. 모델 학습
def fit_model(model, train_inputs, train_labels, val_inputs, val_labels, epochs=100, batch_size=4):
    history = model.fit(
        train_inputs,
        train_labels,
        validation_data=(val_inputs, val_labels),
        epochs=epochs,
        batch_size=batch_size
    )
    return history


def plot_training_history(history):
    """
    Plots training and validation loss over epochs from a model's training history.
    
    Parameters:
    - history: The history object returned by model.fit(), containing training and validation loss.
    """
    # Extract loss values for each epoch
    epochs = range(1, len(history.history['loss']) + 1)
    train_loss = history.history['loss']
    val_loss = history.history['val_loss']

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, train_loss, 'r-', label='Train Loss')  # Red for training loss
    plt.plot(epochs, val_loss, 'b-', label='Validation Loss')  # Blue for validation loss
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.grid(True)
    plt.show()


# 6. 메인 실행 함수
def main(file_path):
    # 설정
    features = [
        "Memory", "Language", "Orientation",
        "Judgment and Problem Solving", "Social Activities", "Home and Hobbies",
        "Daily Living", "Personality and Behavior"
    ]

    # 데이터 전처리
    train_df, val_df, num_classes, label_encoder = preprocess(file_path, features)

    # BERT 모델 및 토크나이저 초기화
    tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
    bert_model = TFDistilBertModel.from_pretrained('distilbert-base-uncased')
    #bert_model.trainable = False  # BERT 모델의 파라미터를 고정
    
    # 인코딩 생성
    train_encodings = tokenize_features(train_df, features, tokenizer)
    val_encodings = tokenize_features(val_df, features, tokenizer)

    # 레이블 준비
    train_labels = tf.convert_to_tensor(train_df['label_encoded'], dtype=tf.int32)
    val_labels = tf.convert_to_tensor(val_df['label_encoded'], dtype=tf.int32)

    # 모델 정의 및 컴파일
    model = create_model(features, num_classes, bert_model)
    model.summary()

    # 입력 데이터 준비
    train_inputs = {}
    val_inputs = {}
    for feature in features:
        feature_key = feature.replace(" ", "_")
        train_inputs[f"{feature_key}_input_ids"] = train_encodings[feature][0]
        train_inputs[f"{feature_key}_attention_mask"] = train_encodings[feature][1]
        val_inputs[f"{feature_key}_input_ids"] = val_encodings[feature][0]
        val_inputs[f"{feature_key}_attention_mask"] = val_encodings[feature][1]

    # 모델 학습
    history = fit_model(model, train_inputs, train_labels, val_inputs, val_labels)
    plot_training_history(history)

    return model, history


# 실행
# file_path에 CSV 파일 경로를 입력하세요.
model, history = main("template.csv")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.llms import Ollama
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.schema import Document  # Document 객체를 가져옵니다.

# 텍스트 변수 선언
text = """<주호소> 
2020.07에 집 비밀번호가 기억이 안났다. 2차례
올해 들어서는 더 기억이 덜 해진 듯 하다
집 비밀번호 기억이 안났다(작년, 올해 1번씩)-나중에도 기억을 못했다
검사 예약날도 어제인줄 알고 보건소 어제 왔었다
금일 당뇨약 챙겨온다는게 못 챙겨왔다 
물건 가지러 갔다 생각안난다 (1주일 1회)
약을 3달치 타면 10개 정도 남는다

<병력>
고혈압(-) 당뇨 (+)
복용약물 : 당뇨약
수술 : 한쪽 난소제거(30년전)
뇌졸중 :  -
두부외상(+) : 10년전 넘어져서 의식소실 1~2분? 정도 MRI 검사 후 이상 없다함
시력 :  -      청력 : -
식사 :  -   수면 : -
술 :   -     담배 : -


<기억력>0.5 MMSE= 26       
주관적 기억력저하?   Y   
증상시작시기?  2020.07
증거?  집 비밀번호 기억이 안났다
증상진행패턴,시작패턴?서서히 진행  
기억력 장애?   Y                 
증상시작시기?  2020.07
증거?  집 비밀번호 기억이 안났다
증상진행패턴,시작패턴?서서히 진행 
현재 일상생활 지장?없다  

단기 기억력: 경미한 장애 
장기 기억력: 정상
물건 잊어버리는 경우?  없음
쇼핑목록 메모 없이 구입할 수 있나? 3가지     
대화내용 이나 10가지 이야기중 기억할 수 있는 개수? 반정도 기억한다
약속은?  적어놓고 잘 지킨다 
어제 저녁식사메뉴기억? 된장찌개, 우거지,고구마 (기억함)
오늘 아침식사메뉴기억? 된장찌개, 우거지,(기억함)
 밥 먹는 것을 잊어버려 또 먹거나 안 먹는 경우(-) 
최근 뉴스/어제 본 뉴스?   요소수 단어는 모르고 내용은 기억함, 날씨 기억함
TV시청 프로그램?  음악에 관한 강의, 섹스폰 배우고 있다
매일 먹는 약은 스스로 잘 챙겨먹음- 3개월치 타면 10개 정도 남는다
전기불, 가스불, 물 잘끄나?없음
냄비 태우는 경우? 없음
자식  1명있음(이름 확인됨) 김미란  사위:류호완
손주  2명있음(이름 확인됨) 류재하,재은
역대대통령? (기억나는대로) 박정희-이승만-노무현     
현 대통령? 문재인
집주소/전화번호기억? 천안시 동남구 원거리 6길 5, 용진빌 401호 ,010-9127-0059

<언어> 증상시작시기?   
의사소통능력: 정상
통상적 수준의 대화 가능? Y           단순한 명령 이해? Y
유창성문제? 어휘력(-), 문법(-), 발음(-)   WFD(-)     Naming diff(-)


<지남력>0
장소 지남력: 정상
MMSE의 시간: 당일날짜? 2021년/  11월/ 11일/ 목요일 /초가을       (+/+/+/+/+)
주요명절, 가족 경조사(기념일/생일) 공과금 납부일? 배우자가 함, 
MMSE의 장소:  충청남도/   천안시/  문화동/  보건소 6층/   (+/+/+/+/+)
 →스스로 집을? 찾음 (자가운전함)
자주 방문하지 않는 낯선 곳? 찾아감 
자주 다니는 익숙한 곳 찾아감   
집안 화장실이나 안방의 위치와 같은 익숙한 실내공간 구분함 
좌우 구분, 방향감각 알수있음     
사람: 알아봄

<판단 및 문제해결> 0
판단력: 정상
숫자계산: MMSE의 100-7? 93-86-79-72-65    (+/+/+/+/+)
종합적 상황 판단력:  정상
사회적 판단력: 예의에 어긋나거나 도덕적행동에 문제없는 것으로 보아 문제없어보임 
MMSE의 옷은 왜 빨아서 입나?   Y      ‘티끌모아태산’의미? Y

<사회활동> 0
직장생활 :  전업주부
친교활동 :  색스폰 배우러 다님
종교활동 :  (변화없음)
장보기 :   (변화없음)

<가정 및 취미활동>0 
일상적인 가사활동: 가전제품 사용 (전기밥솥, 가스렌지, 청소기 , 세탁기 사용함) 리모콘 사용( 전부사용가능 ) 전화 걸기 (스마트폰 ,인터넷 미숙, 문자 카톡사용, 버튼사용  ), 전화 받기(사용가능 ) 
집안일: 잘함 
이전에 즐기던 모든 취미활동:  (변화없음)
은행업무:  잘함
운동:  걷기 매일 1시간


<개인일상생활> 0  
1) 식사: 배고플 때 스스로? (O)    적절한 도구와 양념 사용? (O)     적절한 속도와 매너인가? (O)
→ 숟가락 ,젓가락사용하여 반찬과 밥을 골고루 먹음  
2) 의복: 옷 입고 벗기, 계절에 맞게? (O)   바른 순서(속옷, 겉옷, 신발)? (O)    단추·지퍼 사용? (O)
→옷을 혼자 입고 벗고가능함  
3) 배설: 필요할 때 화장실 사용? (O)      혼자 화장실 사용? (O)      실금? (X)
→대소변조절가능하며 뒤처리도 스스로함 
4) 위생: 세수/ 양치·틀니관리 /머리감기·빗질 /목욕 
 →목욕, 세수, 양치, 샤워 스스로함


<성격, 행동>  GDS=5      (요즘 우울한거 물어보니 성격이 외골수적이라 그런 것 같다
                          자아실현이 안되서 불안하다, 자살 생각도 1주 1~2회 든다한다)

1) 우울증 : 없음(-),일 이상 또는 하루 중 절반이상(+), 거의매일 그리고 하루중 대부분(++)
 
우울(+) 흥미소실(-), 식욕저하(-)또는 체중변화(-), 수면장애(-): 불면 또는 과수면(-),지체 또는 초조(+), 피로 또는 기력 저하(-), 무가치감 또는 죄책감(+), 집중력저하 또는 결단력 저하 (-), 자살사고 또는 시도(+)

위의 증상으로 일상생활에 지장이 있나? 없음

2) 문제 행동 : 없음
"""

# Function to load, split, and retrieve documents from text data
def load_and_retrieve_from_text(text):
    # Document 객체를 사용하여 docs 리스트를 생성합니다.
    docs = [Document(page_content=text)]
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    splits = text_splitter.split_documents(docs)
    embeddings = OllamaEmbeddings(model="mxbai-embed-large")
    vectorstore = Chroma.from_documents(documents=splits, embedding=embeddings)
    return vectorstore.as_retriever()

# Function to format documents
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

# Function that defines the RAG chain using Ollama LLM
def rag_chain_from_text(question):
    retriever = load_and_retrieve_from_text(text)
    retrieved_docs = retriever.invoke(question)
    formatted_context = format_docs(retrieved_docs)
    
    # Ollama LLM을 통해 답변 생성
    prompt_template = """
    Please answer the patient's literacy and numeracy in the following formatted answer.
    ex1) Able to read and write, can perform basic math.
    ex2) Cannot read or write, mathmatic also.  

    Question: {question}
    Context: {context}

    Answer:"""
    prompt = PromptTemplate(input_variables=["question", "context"], template=prompt_template)
    llm = Ollama(model="llama3.2:3b")  # Ollama 모델 사용
    chain = LLMChain(llm=llm, prompt=prompt)
    
    # 질문과 문맥을 전달하여 답변 생성
    answer = chain.run({"question": question, "context": formatted_context})
    return answer

# 예제 질문
question1 = "What are the main complaints of the patient?"
response1 = rag_chain_from_text(question1)
print(response1)
```
